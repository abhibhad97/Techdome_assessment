```python
import pandas as pd
import numpy as np
from textblob import TextBlob
from langdetect import detect
from nltk.util import ngrams
from string import punctuation
import contractions
import nltk
from nltk.corpus import stopwords
from unidecode import unidecode
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer 
from wordcloud import WordCloud
import warnings 
warnings.filterwarnings('ignore')
```

```python
df=pd.read_excel("worksheet.xlsx")
df
```

```python
def lang_detct(data):
    data_lang= detect(data)
    return data_lang
```

```python
df['Article'].apply(lang_detct).unique()
```

```python
df
```

```python
stopword_list= stopwords.words('english')
stopword_list.remove('no')
stopword_list.remove('nor')
stopword_list.remove('not')
print(stopword_list)
```

```python
def remove_newline(data):
    data= data.replace('/n','').replace('/t','')
    return data

def contraction(data):
    data= contractions.fix(data)
    return data
def accented_char(data):
    data= unidecode(data)
    return data

def tokenize(data):
    tokens= word_tokenize(data)
    clean_text = [word.lower() for word in tokens if (word not in punctuation) 
                  and (word.lower() not in stopword_list) and (len(word)>2) and (word.isalpha())]
    return clean_text


def lemmatization(data):
    lem_word= WordNetLemmatizer()
    final_token=[]
    for i in data:
        lemma_word= lem_word.lemmatize(i)
        final_token.append(i)
    return " ".join(final_token)
```

```python
clean_data = df.Article.apply(remove_newline)
clean_text = clean_data.apply(contraction)
clean_text=  clean_text.apply(accented_char)
clean_text = clean_text.apply(tokenize)
clean_text=  clean_text.apply(lemmatization)
clean_text
```

```python
type(clean_text)
df['clean_text']=clean_text
df
```

```python
def analyze_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return 'Positive'
    elif analysis.sentiment.polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'


df['Sentiment'] = df['clean_text'].apply(analyze_sentiment)
df
```

```python
def extract_ngram(data,ngram_range):
    tokens= word_tokenize(data)
    final_ngram= ngrams(tokens,ngram_range)
    final=[]
    
    for i in final_ngram:
        final.append(" ".join([i[0],i[1],i[2]]))
        
    return final

trigram= clean_text.apply(lambda x: extract_ngram(x,3))
fianl_trigram=[]

for i in trigram:
    fianl_trigram.extend(i)
fianl_trigram
```

```python
model.save('w2vec_model.save')
```

```python
def w2vec(data,model):
    reviw_vect=[]
    for review in data:
        zero_vect= np.zeros(model.vector_size)
        word_vect=[]
        for word in review:
            if word in model.wv:
                try:
                    word_vectors= model.wv[word]
                    word_vect.append(word_vectors)
                except keyError:
                    continue
        if word_vect:
            vect_array= np.array(word_vect)
            avg_vect= np.mean(vect_array,axis=0)
            reviw_vect.append(avg_vect)
        else:
            reviw_vect.append(zero_vect)
            
    return reviw_vect
```

```python
word_embd= w2vec(sent,model)
(word_embd)
```

```python
# wordcloud on clean text
def wordcloud(data,column):
    
    df_ = data[column].str.cat(sep=" ")
    wordcloud = WordCloud(width=700,height=500,background_color='white').generate(df_)
    plt.figure(figsize=(10,16))
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()
wordcloud(df,"clean_text")
```

```python
from rake_nltk import Rake
```

```python
def rake_extractor(data, n=50):
    keyword_extractor = Rake()
    keyword_extractor.extract_keywords_from_text(data)
    return [phrase for score, phrase in keyword_extractor.get_ranked_phrases_with_scores()[:n]]

df['Keywords'] = df['clean_text'].apply(rake_extractor)


keylist = [keyword for keyword_list in df['Keywords'] for keyword in keyword_list]

keywords = " ".join(keylist)

wordcloud = WordCloud(width=800, height=400, background_color="white").generate(keywords)

plt.figure(figsize=(10,16))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
